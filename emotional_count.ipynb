```python
import spacy
from spacy.lang.de import German
from transformers import pipeline
```

```python
!python -m spacy download de_core_news_sm
nlp = spacy.load("de_core_news_sm")
```

```python
def all_adjectives(my_string):
    doc = nlp(my_string)
    adjectives = [] # array
    for token in doc:
        if token.pos_ == "ADJ":
            adjectives.append(token.text) # alle token im array ablegen
    return adjectives
```

```python
def how_emotional(adjectives):
    sentiment_analyser = pipeline('sentiment-analysis', model='oliverguhr/german-sentiment-bert')
    counts = {"positive": 0, "negative": 0, "neutral": 0}

    for ad in adjectives:
        classification = sentiment_analyser(ad)[0]['label']
        if classification in counts:
            counts[classification] += 1
        else:
            counts['neutral'] += 1

    return counts
```
