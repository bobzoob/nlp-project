```python
# alle Reden der AfD extrahieren, um nur mit diesen weiter zu arbeiten:

afd = opendiscourse.query('factionId == 0')
print(afd)
```

```python
print(len(afd))
```

```python
for i in range(len(afd.values)):
    speech = afd["speechContent"][i]
    if isinstance(speech, str) is not True:
        print(i, speech)
        
        
# ergibt KeyError 0, ich glaube das heißt, dass hier keine leeren Reden sind?
```

```python
# Index der nur-AfD-Reden resetten:

afd.reset_index(drop=True)
```

```python
# Corpus mit den 100 letzten/aktuellsten Reden erstellen:

extra_stop = ["herr", "präsident", "frau", "dame", "damen", "herr", "herren", "\xa0", " "] 
full_corpus = [] # for the final tokenized corpus

for i in range(4337,4437): # letzte 100 Einträge, da am aktuellsten
       
    speech = afd['speechContent'].iloc[i].replace('\n', ' ') # iterate through column 'speechContent' on i. row 
    # and replace all trailing new-line commands there with whitespace
     
    tokenized_speech = nlp(speech) # feed to our German language model
    
    speech_corpus = []
    for t in tokenized_speech:
        if not t.is_stop and not t.is_punct and not t.like_num and t.lemma_.lower() not in extra_stop:
            speech_corpus.append(t.lemma_.lower())

    full_corpus.append(speech_corpus)
```
